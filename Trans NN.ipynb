```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df = df.drop(['ID_code'],axis = 1)
```

```python
X = df.drop(['target'],axis = 1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
df.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
def get_model():
    model = Sequential()
    model.add(Dense(100,input_dim = 200,activation = 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(150,activation = 'relu'))
    model.add(Dense(228,activation = 'relu'))
    model.add(Dense(118,activation = 'relu'))
    model.add(Dense(1,activation = 'sigmoid'))
    model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
    return model
```

```python
model = get_model()
```

```python
model.fit(X_train,y_train,epochs = 300,batch_size = 5789, validation_data = (X_test,y_test))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metrics_df = pd.DataFrame(model.history.history)
metrics_df.plot(grid = 1)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import roc_auc_score
```

```python
y_pred = model.predict(X_test)
```

```python
fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_test, y_pred)
auc_keras = auc(fpr_keras, tpr_keras)
roc_auc_score(y_test, y_pred)
```

```python
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_keras, tpr_keras, label='Keras (area = {:.3f})'.format(auc_keras))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```
